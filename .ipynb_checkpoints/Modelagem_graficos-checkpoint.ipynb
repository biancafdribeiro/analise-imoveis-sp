# 📊 Análise Visual — Imóveis SP

Nove gráficos salvos em `.jpg` cobrindo correlação, distribuição, comparação de modelos, resíduos, features e análise exploratória.

```python
# ── Estilo global claro ──────────────────────────────────────────
plt.rcParams.update({
    'figure.facecolor': 'white', 'axes.facecolor': '#f8f9fb',
    'axes.edgecolor': '#d0d4de', 'axes.labelcolor': '#1a1d2e',
    'axes.titlecolor': '#1a1d2e', 'axes.titlesize': 13, 'axes.labelsize': 11,
    'xtick.color': '#444860', 'ytick.color': '#444860',
    'text.color': '#1a1d2e', 'grid.color': '#e2e5ef', 'grid.linewidth': 0.7,
    'legend.facecolor': 'white', 'legend.edgecolor': '#d0d4de',
})
C1, C2, C3, C4, C5 = '#5b5ef4','#0ea5e9','#10b981','#f97316','#ec4899'
PAL = [C1, C2, C3, C4, C5]
```

```python
# ── Estilo global dark ────────────────────────────────────────────
plt.rcParams.update({
    'figure.facecolor': '#0f1117', 'axes.facecolor': '#1a1d27',
    'axes.edgecolor': '#2e3250', 'axes.labelcolor': '#c9d1d9',
    'axes.titlecolor': '#e6edf3', 'axes.titlesize': 13, 'axes.labelsize': 11,
    'xtick.color': '#8b949e', 'ytick.color': '#8b949e',
    'text.color': '#c9d1d9', 'grid.color': '#21262d', 'grid.linewidth': 0.6,
    'legend.facecolor': '#161b22', 'legend.edgecolor': '#30363d',
})
ACCENT, ACCENT2, ACCENT3, ACCENT4, ACCENT5 = '#7c6af7','#38bdf8','#34d399','#f97316','#f472b6'
PALETTE = [ACCENT, ACCENT2, ACCENT3, ACCENT4, ACCENT5]
```

```python
# ── Pipeline base ─────────────────────────────────────────────────
df = pd.read_csv('imoveis_sp_limpo.csv', encoding='utf-8-sig')
df = df.dropna()
df = df[(df['preco_venda'] >= df['preco_venda'].quantile(0.01)) &
        (df['preco_venda'] <= df['preco_venda'].quantile(0.99))]

X = df.drop('preco_venda', axis=1)
y = np.log1p(df['preco_venda'])
X_enc = pd.get_dummies(X, columns=['bairro', 'tipo_imovel'], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.2, random_state=42)

rf = RandomForestRegressor(n_estimators=200, max_depth=None, min_samples_leaf=2,
                           max_features='sqrt', random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)
print(f'R² Random Forest: {r2_score(y_test, pred_rf):.4f}')
```

## 1 · Matriz de Correlação

```python
num_cols = ['area_util','banheiros','suites','quartos','vagas_garagem','preco_venda']
labels   = ['Área Útil','Banheiros','Suítes','Quartos','Vagas','Preço Venda']
corr = df[num_cols].corr()
corr.index = labels; corr.columns = labels

fig, ax = plt.subplots(figsize=(9, 7))
fig.patch.set_facecolor('white'); ax.set_facecolor('white')
sns.heatmap(corr, mask=None, cmap='RdYlGn', center=0, vmin=-1, vmax=1,
            annot=True, fmt='.2f', linewidths=1, linecolor='white',
            square=True, annot_kws={'size':11,'color':'#1a1d2e','weight':'bold'},
            ax=ax, cbar_kws={'shrink':0.8,'label':'Correlação de Pearson'})
ax.set_title('Matriz de Correlação — Variáveis Numéricas', pad=15, fontweight='bold', fontsize=14)
ax.tick_params(axis='x', rotation=30, labelsize=10)
ax.tick_params(axis='y', rotation=0, labelsize=10)
plt.tight_layout()
plt.savefig('plot1_correlacao.jpg', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
```

## 2 · Distribuição do Preço: Original vs Log

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.patch.set_facecolor('#0f1117')
fig.suptitle('Distribuição do Preço de Venda', fontsize=14, fontweight='bold', y=1.01)
for ax in axes: ax.set_facecolor('#1a1d27'); ax.grid(True, axis='y', alpha=0.3)

axes[0].hist(df['preco_venda']/1e6, bins=80, color=ACCENT, alpha=0.85, edgecolor='none')
axes[0].set_title('Escala Original (R$ milhões)')
axes[0].set_xlabel('Preço (R$ mi)'); axes[0].set_ylabel('Frequência')
axes[0].axvline(df['preco_venda'].median()/1e6, color=ACCENT2, lw=1.8, ls='--',
                label=f'Mediana: R${df["preco_venda"].median()/1e6:.2f}mi')
axes[0].legend()

axes[1].hist(np.log1p(df['preco_venda']), bins=80, color=ACCENT3, alpha=0.85, edgecolor='none')
axes[1].set_title('Escala Log1p (normalizada)')
axes[1].set_xlabel('log(Preço + 1)'); axes[1].set_ylabel('Frequência')
axes[1].axvline(np.log1p(df['preco_venda']).median(), color=ACCENT2, lw=1.8, ls='--',
                label=f'Mediana: {np.log1p(df["preco_venda"]).median():.2f}')
axes[1].legend()
plt.tight_layout()
plt.savefig('plot2_distribuicao_preco.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 3 · Comparação de Modelos

```python
models = {
    'Random Forest\n(tuned)': rf,
    'Decision Tree': DecisionTreeRegressor(max_depth=15, random_state=42),
    'Ridge\nRegression': Ridge(alpha=1.0),
    'Gradient\nBoosting': GradientBoostingRegressor(n_estimators=150, max_depth=5, random_state=42, subsample=0.8),
}
results = {}
for name, model in models.items():
    if name != 'Random Forest\n(tuned)': model.fit(X_train, y_train)
    p = model.predict(X_test)
    p_r, y_r = np.expm1(p), np.expm1(y_test)
    results[name] = {'R² (log)': r2_score(y_test, p),
                     'MAE (R$k)': mean_absolute_error(y_r, p_r)/1000,
                     'RMSE (R$k)': np.sqrt(mean_squared_error(y_r, p_r))/1000}
res_df = pd.DataFrame(results).T.reset_index().rename(columns={'index':'Modelo'})

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.patch.set_facecolor('#0f1117')
fig.suptitle('Comparação de Modelos', fontsize=14, fontweight='bold', y=1.02)
for ax, metric, color, b in zip(axes, ['R² (log)','MAE (R$k)','RMSE (R$k)'],
                                  [ACCENT, ACCENT4, ACCENT5], ['max','min','min']):
    ax.set_facecolor('#1a1d27')
    vals = res_df[metric].values
    bars = ax.barh(res_df['Modelo'], vals, color=color, alpha=0.85, edgecolor='none', height=0.55)
    ax.set_title(metric, fontweight='bold'); ax.grid(True, axis='x', alpha=0.3)
    best_idx = int(np.argmax(vals) if b=='max' else np.argmin(vals))
    bars[best_idx].set_color(ACCENT3); bars[best_idx].set_alpha(1.0)
    for bar, v in zip(bars, vals):
        ax.text(v+max(vals)*0.01, bar.get_y()+bar.get_height()/2,
                f'{v:.3f}' if metric.startswith('R') else f'{v:.0f}',
                va='center', color='#e6edf3', fontsize=9)
plt.tight_layout()
plt.savefig('plot3_comparacao_modelos.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 4 · Real vs Predito (Densidade)

```python
y_real = np.expm1(y_test.values); p_real = np.expm1(pred_rf)
lim_max = np.percentile(y_real, 99)

fig, ax = plt.subplots(figsize=(8, 8))
fig.patch.set_facecolor('#0f1117'); ax.set_facecolor('#1a1d27')
hb = ax.hexbin(y_real/1e6, p_real/1e6, gridsize=60, cmap='plasma', mincnt=1,
               extent=[0,lim_max/1e6,0,lim_max/1e6], linewidths=0.1)
cb = fig.colorbar(hb, ax=ax, label='Contagem')
cb.ax.yaxis.label.set_color('#c9d1d9'); cb.ax.tick_params(colors='#8b949e')
lim = lim_max/1e6
ax.plot([0,lim],[0,lim],'--',color=ACCENT3, lw=1.8, label='Predição perfeita')
ax.set_xlim(0,lim); ax.set_ylim(0,lim)
ax.set_xlabel('Preço Real (R$ mi)'); ax.set_ylabel('Preço Predito (R$ mi)')
ax.set_title(f'Real vs Predito — Random Forest\nR² = {r2_score(y_test, pred_rf):.4f}  |  MAE = R${mean_absolute_error(y_real,p_real)/1000:.0f}k',
             fontweight='bold')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plot4_real_vs_predito.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 5 · Análise de Resíduos

```python
residuos = y_test.values - pred_rf
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.patch.set_facecolor('#0f1117')
fig.suptitle('Análise de Resíduos — Random Forest', fontsize=14, fontweight='bold', y=1.01)
for ax in axes: ax.set_facecolor('#1a1d27'); ax.grid(True, alpha=0.3)

axes[0].scatter(pred_rf, residuos, alpha=0.15, s=5, color=ACCENT, rasterized=True)
axes[0].axhline(0, color=ACCENT3, lw=1.8, ls='--')
axes[0].set_xlabel('Valor Predito (log)'); axes[0].set_ylabel('Resíduo')
axes[0].set_title('Resíduos vs Valor Predito')

axes[1].hist(residuos, bins=80, color=ACCENT2, alpha=0.8, edgecolor='none', density=True)
xr = np.linspace(residuos.min(), residuos.max(), 300)
axes[1].plot(xr, sp_stats.norm.pdf(xr, residuos.mean(), residuos.std()),
             color=ACCENT4, lw=2, label='Normal teórica')
axes[1].set_xlabel('Resíduo'); axes[1].set_ylabel('Densidade')
axes[1].set_title('Distribuição dos Resíduos'); axes[1].legend()
plt.tight_layout()
plt.savefig('plot5_residuos.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 6 · Top 20 Features — Importância

```python
importances = pd.Series(rf.feature_importances_, index=X_enc.columns)
top20 = importances.sort_values(ascending=True).tail(20)

fig, ax = plt.subplots(figsize=(10, 7))
fig.patch.set_facecolor('#0f1117'); ax.set_facecolor('#1a1d27')
clrs = [ACCENT3 if v > top20.quantile(0.7) else ACCENT for v in top20.values]
bars = ax.barh(top20.index, top20.values, color=clrs, alpha=0.9, edgecolor='none', height=0.7)
ax.set_title('Top 20 Features — Importância (Random Forest)', fontweight='bold')
ax.set_xlabel('Importância (Gini)'); ax.grid(True, axis='x', alpha=0.3)
for bar, v in zip(bars, top20.values):
    ax.text(v+0.0003, bar.get_y()+bar.get_height()/2, f'{v:.4f}', va='center', color='#c9d1d9', fontsize=8)
plt.tight_layout()
plt.savefig('plot6_importancia_features.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 7 · Preço por Tipo de Imóvel

```python
tipo_counts = df['tipo_imovel'].value_counts()
tipos_top = tipo_counts[tipo_counts > 200].index.tolist()
df_tipo = df[df['tipo_imovel'].isin(tipos_top)]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.patch.set_facecolor('#0f1117')
fig.suptitle('Preço por Tipo de Imóvel', fontsize=14, fontweight='bold', y=1.01)
med = df_tipo.groupby('tipo_imovel')['preco_venda'].median().sort_values(ascending=True)/1e6
clrs = [PALETTE[i%len(PALETTE)] for i in range(len(med))]

axes[0].set_facecolor('#1a1d27'); axes[0].grid(True, axis='x', alpha=0.3)
axes[0].barh(med.index, med.values, color=clrs, alpha=0.85, edgecolor='none', height=0.6)
axes[0].set_xlabel('Preço Mediano (R$ mi)'); axes[0].set_title('Mediana por Tipo')
for i,(idx,v) in enumerate(med.items()):
    axes[0].text(v+0.01, i, f'R${v:.2f}mi', va='center', color='#c9d1d9', fontsize=8)

axes[1].set_facecolor('#1a1d27'); axes[1].grid(True, axis='x', alpha=0.3)
data_box = [df_tipo[df_tipo['tipo_imovel']==t]['preco_venda'].values/1e6 for t in med.index]
bp = axes[1].boxplot(data_box, vert=False, patch_artist=True, labels=med.index,
                     medianprops={'color':ACCENT3,'lw':2}, whiskerprops={'color':'#8b949e'},
                     capprops={'color':'#8b949e'},
                     flierprops={'marker':'o','markersize':2,'alpha':0.3,'color':'#8b949e'})
for patch, c in zip(bp['boxes'], clrs): patch.set_facecolor(c); patch.set_alpha(0.6)
axes[1].set_xlabel('Preço (R$ mi)'); axes[1].set_title('Boxplot por Tipo')
plt.tight_layout()
plt.savefig('plot7_preco_por_tipo.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 8 · Top 15 Bairros por Preço Mediano

```python
bairro_med = df.groupby('bairro')['preco_venda'].agg(['median','count'])
bairro_med = bairro_med[bairro_med['count']>=30].sort_values('median',ascending=False).head(15)
bairro_med['median'] = bairro_med['median']/1e6

fig, ax = plt.subplots(figsize=(11, 7))
fig.patch.set_facecolor('#0f1117'); ax.set_facecolor('#1a1d27'); ax.grid(True, axis='x', alpha=0.3)
clr_b = [ACCENT if i<5 else (ACCENT2 if i<10 else ACCENT3) for i in range(len(bairro_med))]
bars = ax.barh(bairro_med.index[::-1], bairro_med['median'][::-1], color=clr_b[::-1], alpha=0.85, edgecolor='none', height=0.65)
ax.set_xlabel('Preço Mediano (R$ mi)')
ax.set_title('Top 15 Bairros — Maior Preço Mediano\n(mín. 30 imóveis)', fontweight='bold')
for bar, v in zip(bars, bairro_med['median'][::-1].values):
    ax.text(v+0.01, bar.get_y()+bar.get_height()/2, f'R${v:.2f}mi', va='center', color='#c9d1d9', fontsize=9)
plt.tight_layout()
plt.savefig('plot8_top_bairros.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```

## 9 · Preço vs Área Útil

```python
sample = df.sample(3000, random_state=42)
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('#0f1117'); ax.set_facecolor('#1a1d27'); ax.grid(True, alpha=0.25)
sc = ax.scatter(sample['area_util'], sample['preco_venda']/1e6,
                c=sample['quartos'], cmap='plasma', alpha=0.5, s=15, edgecolors='none')
cb = fig.colorbar(sc, ax=ax, label='Nº de Quartos')
cb.ax.yaxis.label.set_color('#c9d1d9'); cb.ax.tick_params(colors='#8b949e')
z = np.polyfit(sample['area_util'], sample['preco_venda']/1e6, 1)
xs = np.linspace(sample['area_util'].min(), sample['area_util'].max(), 200)
ax.plot(xs, np.poly1d(z)(xs), color=ACCENT3, lw=2, ls='--', label='Tendência linear')
r, _ = sp_stats.pearsonr(sample['area_util'], sample['preco_venda'])
ax.set_title(f'Preço vs Área Útil  (r = {r:.3f})', fontweight='bold')
ax.set_xlabel('Área Útil (m²)'); ax.set_ylabel('Preço (R$ mi)'); ax.legend()
plt.tight_layout()
plt.savefig('plot9_preco_vs_area.jpg', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
```
