```python
#Foram mantidos apenas imóveis residenciais (apartamentos, casas, coberturas, flats, kitnets e imóveis em condomínio), 
#pois imóveis comerciais, terrenos e propriedades rurais possuem dinâmicas de precificação distintas, o que poderia 
#introduzir ruído e prejudicar o desempenho dos modelos preditivos.

tipos_residenciais = [
    'Apartamento',
    'Casa',
    'Casa de dois andares',
    'Condomínio',
    'Cobertura',
    'Flat',
    'Kitnet'
]

df_res = df_venda[
    df_venda['tipo_imovel'].isin(tipos_residenciais)
].copy()
print(df_res.shape)
```
