# 🏙️ Simulador de Preço — Imóveis SP

Notebook de aplicação autossuficiente. Carrega o modelo salvo pelo `Modelagem.ipynb` e exibe um simulador interativo.

> **Pré-requisito:** execute `Modelagem.ipynb` até a célula de `joblib.dump` para gerar os arquivos `.pkl` em `../data/`.

```python
import pandas as pd
import numpy as np
import joblib
import ipywidgets as widgets
from IPython.display import display, clear_output
```

## 1 · Carrega modelo e dados

```python
# Modelo e colunas salvas pelo Modelagem.ipynb
rf             = joblib.load('../data/random_forest_model.pkl')
colunas_modelo = joblib.load('../data/colunas_modelo.pkl')

# CSV para popular dropdowns e calcular mediana por bairro
df = pd.read_csv('../data/imoveis_sp_limpo.csv', encoding='utf-8-sig')
df = df.dropna()
df = df[(df['preco_venda'] >= df['preco_venda'].quantile(0.01)) &
        (df['preco_venda'] <= df['preco_venda'].quantile(0.99))]

bairros = sorted(df['bairro'].unique())
tipos   = sorted(df['tipo_imovel'].unique())

print(f'Modelo carregado | {len(colunas_modelo)} features')
print(f'Base: {len(df):,} imóveis | {len(bairros)} bairros | {len(tipos)} tipos')
```

## 2 · Simulador interativo

Ajuste os controles — o preço atualiza automaticamente a cada mudança.

```python
# ── Widgets de entrada ───────────────────────────────────────────
style  = {'description_width': '130px'}
w400   = widgets.Layout(width='420px')
w500   = widgets.Layout(width='520px')

w_bairro = widgets.Dropdown(
    options=bairros,
    value='Moema' if 'Moema' in bairros else bairros[0],
    description='📍 Bairro:',
    style=style, layout=w400
)
w_tipo = widgets.Dropdown(
    options=tipos,
    value='Apartamento' if 'Apartamento' in tipos else tipos[0],
    description='🏠 Tipo:',
    style=style, layout=w400
)
w_area = widgets.IntSlider(
    min=20, max=750, step=10, value=100,
    description='📐 Área (m²):',
    style=style, layout=w500
)
w_quartos = widgets.IntSlider(
    min=0, max=10, value=2,
    description='🛏️ Quartos:',
    style=style, layout=w400
)
w_banheiros = widgets.IntSlider(
    min=0, max=10, value=2,
    description='🚿 Banheiros:',
    style=style, layout=w400
)
w_suites = widgets.IntSlider(
    min=0, max=10, value=1,
    description='🛁 Suítes:',
    style=style, layout=w400
)
w_vagas = widgets.IntSlider(
    min=0, max=10, value=1,
    description='🚗 Vagas:',
    style=style, layout=w400
)

out = widgets.Output()

# ── Função de previsão ────────────────────────────────────────────
def prever(change=None):
    novo = pd.DataFrame([{
        'bairro':        w_bairro.value,
        'tipo_imovel':   w_tipo.value,
        'area_util':     float(w_area.value),
        'quartos':       float(w_quartos.value),
        'banheiros':     float(w_banheiros.value),
        'suites':        float(w_suites.value),
        'vagas_garagem': float(w_vagas.value),
    }])

    # Mesmo encoding do treino
    novo_enc = pd.get_dummies(novo, columns=['bairro', 'tipo_imovel'])
    novo_enc = novo_enc.reindex(columns=colunas_modelo, fill_value=0)

    pred_log = rf.predict(novo_enc)[0]
    preco    = np.expm1(pred_log)
    low      = preco * 0.85   # intervalo -15%
    high     = preco * 1.15   # intervalo +15%

    # Comparativo com mediana do bairro e tipo selecionados
    filtro     = (df['bairro'] == w_bairro.value) & (df['tipo_imovel'] == w_tipo.value)
    sub        = df[filtro]['preco_venda']
    med_bairro = sub.median() if len(sub) >= 3 else df[df['bairro'] == w_bairro.value]['preco_venda'].median()
    n_imoveis  = len(sub) if len(sub) >= 3 else len(df[df['bairro'] == w_bairro.value])
    delta      = (preco - med_bairro) / med_bairro * 100
    sinal      = '▲' if delta > 0 else '▼'

    with out:
        clear_output(wait=True)
        sep = '─' * 50
        print(sep)
        print(f'  💰  Preço estimado      R$ {preco:>13,.0f}')
        print(f'  📉  Intervalo mínimo    R$ {low:>13,.0f}  (−15%)')
        print(f'  📈  Intervalo máximo    R$ {high:>13,.0f}  (+15%)')
        print(sep)
        print(f'  🏘️   Mediana {w_bairro.value[:18]:<18}  R$ {med_bairro:>10,.0f}')
        print(f'  📊  Sua estimativa vs. bairro   {sinal} {abs(delta):.1f}%')
        print(f'  🔢  Imóveis similares na base    {n_imoveis:,}')
        print(sep)

# Conecta todos os widgets
for w in [w_bairro, w_tipo, w_area, w_quartos, w_banheiros, w_suites, w_vagas]:
    w.observe(prever, names='value')

# ── Layout ────────────────────────────────────────────────────────
titulo = widgets.HTML(
    '<h3 style="font-family:sans-serif;margin:8px 0">'
    '🏙️ Simulador de Preço — Imóveis SP</h3>'
    '<p style="font-family:sans-serif;color:#555;margin:0 0 12px">'
    'Ajuste os parâmetros — o preço atualiza automaticamente.</p>'
)

col_esq = widgets.VBox([w_bairro, w_tipo, w_area])
col_dir = widgets.VBox([w_quartos, w_banheiros, w_suites, w_vagas])
painel  = widgets.HBox(
    [col_esq, col_dir],
    layout=widgets.Layout(gap='32px')
)

display(widgets.VBox([titulo, painel, out]))
prever()  # calcula previsão inicial ao rodar a célula
```
